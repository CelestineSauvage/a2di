## Question 6
Calculez, à partir des données, l’écart-type de la taille chez l’homme et chez la femme.

```python
def moyenne(l):
    return sum(l, 0.0) / len(l)

def variance(l):
    m=moyenne(l)
    return moyenne([(x-m)**2 for x in l])

def ecart_type(l):
    return variance(l)**0.5
```

```python
print("Ecart type femme :" + str(ecart_type(taille_f)))
print("Ecart type homme :" + str(ecart_type(taille_h)))
```

```python
def p_train(X, Y, n):
    w = np.zeros(len(X[0])+1)
    X2 = np.insert(X, len(X[0]), 1, axis = 1)

    for t in range(n):
        for i, x in enumerate(X2):
            y = np.sign(np.dot(w, x))
            if (y != Y[i]) :
                w += x*Y[i]
    return w

def predict(x, w):
    x = np.insert(x, len(x), 1)
    return np.sign(np.dot(w, x))

def p_test(X, w):
    c_hat = []
    for x_test in (X):
        c_hat.append(int(predict(x_test, w)))
    return c_hat
```

```python
def init_points(size):
    pts = np.random.random_sample([size,2])
    y = np.empty([size],dtype=np.int64)
    for i, pt in enumerate(pts): 
        if (((-1/2)*pt[0]) + 0.75 <= pt[1]):
            y[i] = +1
        else :
            y[i] = -1
    return pts,y

def data_gen(size):
    pts,y = init_points(size)
    size_train = int((size * 10) / 100)
    X_train = pts[0:size_train+1]
    c_train = y[0:size_train+1]
    X_test = pts[size_train+1::]
    c_test = y[size_train+1::]
    return X_train, X_test, c_train, c_test
```

```python
X_train,X_test,c_train,c_test = data_gen(200)
w = p_train(X_train, c_train, 30)
c_hat = p_test(X_test, w)
print(c_hat)
print(c_test)
```

```python
def loss(c, y):
    return 0 if (c==y) else 1

def get_test_err(c_hat,c_test):
    err = 0
    n = len(c_hat)
    for i in range(n):
        err += loss(c_hat[i],c_test[i])
    return (err/n)
```
